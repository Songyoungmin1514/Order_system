```python
# ---------------------------------------------
# ë¬´ì¸ ì£¼ë¬¸ ì‹œìŠ¤í…œ V2
# ê¸°ëŠ¥ ì¶”ê°€: ì¥ë°”êµ¬ë‹ˆ ê´€ë¦¬, UI ë©”ë‰´ ì¶”ê°€, ì¤‘ë³µ í•©ì‚° ì²˜ë¦¬
# ---------------------------------------------

menu = {
    1: {"name": "ì•„ë©”ë¦¬ì¹´ë…¸", "price": 2000},
    2: {"name": "ì¹´í˜ë¼ë–¼", "price": 2500},
    3: {"name": "ëŒì²´ë¼ë–¼", "price": 3800},
    4: {"name": "ë”¸ê¸°ë¼ë–¼", "price": 4900}
}

# cartë¥¼ dictë¡œ ë³€ê²½ â†’ {ë©”ë‰´id: ìˆ˜ëŸ‰}
cart = {}


# ----------------- UI -----------------

def show_main_menu():
    """ë©”ì¸ UI ë©”ë‰´ ì¶œë ¥"""
    print("\n==== ë¬´ì¸ ì£¼ë¬¸ ì‹œìŠ¤í…œ ====")
    print("1. ë©”ë‰´ í™•ì¸ & ì£¼ë¬¸ ì¶”ê°€")
    print("2. ì¥ë°”êµ¬ë‹ˆ í™•ì¸/ê´€ë¦¬")
    print("3. ì£¼ë¬¸ ì¢…ë£Œ(ê²°ì œ)")
    print("==========================")

def show_menu(menu: dict):
    """ë©”ë‰´ ì¶œë ¥"""
    print("\nğŸ“Œ ë©”ë‰´íŒ")
    for key, item in menu.items():
        print(f"{key}. {item['name']} - {item['price']}ì›")
    print("-----------------")


# ----------------- ê¸°ëŠ¥ -----------------

def add_to_cart(cart: dict, menu: dict):
    """ë©”ë‰´ ì£¼ë¬¸ ì¶”ê°€ + ì¤‘ë³µì‹œ ìë™ ìˆ˜ëŸ‰ í•©ì‚°"""
    try:
        item_id = int(input("ë©”ë‰´ ë²ˆí˜¸ë¥¼ ì…ë ¥í•˜ì„¸ìš”: "))
        if item_id not in menu:
            print("âŒ ì—†ëŠ” ë©”ë‰´ì…ë‹ˆë‹¤.")
            return cart

        qty = int(input("ìˆ˜ëŸ‰ ì…ë ¥: "))
        if qty <= 0:
            print("âŒ ìˆ˜ëŸ‰ì€ 1 ì´ìƒ")
            return cart

        # ì¤‘ë³µ ì£¼ë¬¸ì‹œ í•©ì‚°
        cart[item_id] = cart.get(item_id, 0) + qty
        print(f"ğŸ›’ '{menu[item_id]['name']}' {qty}ê°œ ì¶”ê°€!")

    except ValueError:
        print("âŒ ìˆ«ìë§Œ ì…ë ¥")
    return cart


def manage_cart(cart: dict, menu: dict):
    """ì¥ë°”êµ¬ë‹ˆ í™•ì¸ ë° ê´€ë¦¬"""
    if not cart:
        print("\nğŸ›’ ì¥ë°”êµ¬ë‹ˆê°€ ë¹„ì—ˆìŠµë‹ˆë‹¤.\n")
        return cart

    print("\n===== ì¥ë°”êµ¬ë‹ˆ =====")
    for idx, (item_id, qty) in enumerate(cart.items(), 1):
        print(f"{idx}. {menu[item_id]['name']} x {qty}")
    print("====================")

    try:
        choice = int(input("ê´€ë¦¬í•  í•­ëª© ë²ˆí˜¸ ì„ íƒ (0=ë’¤ë¡œê°€ê¸°): "))
        if choice == 0:
            return cart
        
        # ì…ë ¥ index â†’ ì‹¤ì œ key ë§¤í•‘
        item_ids = list(cart.keys())
        if choice-1 not in range(len(item_ids)):
            print("âŒ ì˜ëª»ëœ ë²ˆí˜¸")
            return cart

        selected = item_ids[choice-1]

        print("\n1. ìˆ˜ëŸ‰ ë³€ê²½")
        print("2. í•­ëª© ì‚­ì œ")
        action = int(input("ì„ íƒ: "))

        if action == 1:
            cart = update_quantity(cart, selected, menu)
        elif action == 2:
            cart = remove_from_cart(cart, selected, menu)
        else:
            print("âŒ ì„ íƒ ì˜¤ë¥˜")

    except ValueError:
        print("âŒ ìˆ«ì ì…ë ¥ í•„ìš”")

    return cart


def update_quantity(cart: dict, item_id: int, menu: dict):
    """ì„ íƒ í•­ëª© ìˆ˜ëŸ‰ ì§ì ‘ ë³€ê²½"""
    try:
        new_qty = int(input("ìƒˆ ìˆ˜ëŸ‰ ì…ë ¥: "))
        if new_qty <= 0:
            print("âŒ ìˆ˜ëŸ‰ì€ 1 ì´ìƒ")
            return cart

        cart[item_id] = new_qty
        print(f"ğŸ”„ {menu[item_id]['name']} ìˆ˜ëŸ‰ì´ {new_qty}ìœ¼ë¡œ ë³€ê²½ë˜ì—ˆìŠµë‹ˆë‹¤.")

    except ValueError:
        print("âŒ ìˆ«ì ì…ë ¥ í•„ìš”")
    return cart


def remove_from_cart(cart: dict, item_id: int, menu: dict):
    """ì¥ë°”êµ¬ë‹ˆì—ì„œ ì•„ì´í…œ ì‚­ì œ"""
    del cart[item_id]
    print(f"âŒ '{menu[item_id]['name']}' ì‚­ì œ ì™„ë£Œ")
    return cart


def print_receipt(cart: dict, menu: dict):
    """ìµœì¢… ì£¼ë¬¸ ì˜ìˆ˜ì¦"""
    print("\n====== ğŸ§¾ ì£¼ë¬¸ ë‚´ì—­ ======")
    total = 0
    for item_id, qty in cart.items():
        name = menu[item_id]['name']
        price = menu[item_id]['price']
        sum_price = price * qty
        total += sum_price
        print(f"{name} x {qty} = {sum_price}ì›")
    print("---------------------------")
    print(f"ğŸ’° ì´ ê²°ì œ ê¸ˆì•¡: {total}ì›")
    print("===========================\n")


# ----------------- Main Loop -----------------

while True:
    show_main_menu()
    try:
        cmd = int(input("ë²ˆí˜¸ ì„ íƒ: "))
    except ValueError:
        print("âŒ ìˆ«ìë§Œ ì…ë ¥")
        continue

    if cmd == 1:
        show_menu(menu)
        cart = add_to_cart(cart, menu)

    elif cmd == 2:
        cart = manage_cart(cart, menu)

    elif cmd == 3:
        print_receipt(cart, menu)
        print("ğŸµ ì´ìš©í•´ì£¼ì…”ì„œ ê°ì‚¬í•©ë‹ˆë‹¤.")
        break

    else:
        print("âŒ ì—†ëŠ” ë©”ë‰´ì…ë‹ˆë‹¤. ë‹¤ì‹œ ì„ íƒí•´ì£¼ì„¸ìš”.")
```
